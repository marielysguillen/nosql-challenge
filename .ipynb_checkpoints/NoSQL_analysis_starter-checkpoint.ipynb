### 2. Which establishments in London have a `RatingValue` greater than or equal to 4?

```python
# Find the establishments with London as the Local Authority and has a RatingValue greater than or equal to 4.
query2 = {'LocalAuthorityName': {'$regex': 'London'},
         'RatingValue': {'$gte': 4}
        }

LocalAut_ratVal = establishments.find(query2)
#$gte Matches values that are greater than or equal to (≥) a specified value
#$regex “Regular expression” can be used to find specific words in the value of a field
```

```python
# Use count_documents to display the number of documents in the result
establishments.count_documents(query2)
```

```python
# Display the first document in the results using pprint
pprint(establishments.find_one(query2))
```

```python
# Convert the result to a Pandas DataFrame
results2_df = pd.DataFrame(LocalAut_ratVal)

# Display the number of rows in the DataFrame
print(len(results2_df))
```

```python
# Display the first 10 rows of the DataFrame
results2_df.head(10)
```

### 4. How many establishments in each Local Authority area have a hygiene score of 0?

```python
# Create a pipeline that:
# 1. Matches establishments with a hygiene score of 0
match_query = {'$match': {'scores.Hygiene': {'$eq': 0}}}

# 2. Groups the matches by Local Authority
group_query = {'$group': {'_id': "$LocalAuthorityName", 'count': {'$sum': 1}}}

# 3. Sorts the matches from highest to lowest
sort_values = {'$sort': {'count': -1}} #sort by count in Descending order (-1)

# Print the number of documents in the result
# pipeline = [group_query, match_query, sort_values] # Put the pipeline together
pipeline = [match_query, group_query, sort_values]

# Run the pipeline through the aggregate method, cast the results as a list, and save the results to a variable
results_final = list(establishments.aggregate(pipeline))
```

```python
# Print the number of rows in the result_final
print("Number of rows in results_final: ", len(results_final))
```

```python
# Print the first 10 results
pprint(results_final[0:10])
```

```python
# Convert the result to a Pandas DataFrame
results_final_df = pd.DataFrame(results_final)

# Display the number of rows in the DataFrame
print('Number of rows in the DataFrame', len(results_final_df))
```

```python
# Display the first 10 rows of the DataFrame
results_final_df.head(10)
```
